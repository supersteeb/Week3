````python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
````

````python
songs_dataset = pd.read_json('MasterSongList.json')
````

````python
songs_dataset.shape
````

````python
from string import punctuation
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
from nltk.stem.snowball import SnowballStemmer

translator = str.maketrans('', '', punctuation)
stemmer = SnowballStemmer('english')
````

````python
def clean_text(raw_text):
#   initialize empty clean_words list
    clean_words = []
    raw_text = raw_text.translate(translator)
    raw_text_split = raw_text.split()

    for word in raw_text_split:
        if word not in ENGLISH_STOP_WORDS:
            clean_words.append(stemmer.stem(word))
    return ' '.join(clean_words)
````

````python
songs_have_lyrics = songs_dataset[songs_dataset['lyrics_features'].str.len() != 0]
````

````python
#dropping the rows in which column['lyrics_features'] is an empty list — converts to string though...
#songs_dataset_has_lyrics = songs_dataset[songs_dataset['lyrics_features'].str.len() != 0]
````

````python
#songs_dataset_has_lyrics = songs_dataset_has_lyrics.reset_index
````

````python
songs_have_lyrics.shape
````

````python
songs_have_lyrics.reset_index()
````

````python
lyricsFeatureList = []
for lyrics_feature in songs_have_lyrics['lyrics_features']:
    lyricsFeatureList.append(' '.join(lyrics_feature))
````

````python
len(lyricsFeatureList)
````

````python
clean_lyricsFeatureList = []
for song in lyricsFeatureList:
    clean_lyricsFeatureList.append(clean_text(song))
    #print('clean_lyricsFeatureList is getting: ', clean_text(song))
````

````python
sd = songs_have_lyrics.copy()

se = pd.Series(clean_lyricsFeatureList)
sd['cleaned_lyrics'] = se.values
````

````python
sd.to_json('songs_have_lyrics.json')
````

# Bag of Words

````python
count_vectorizer = CountVectorizer()
bag_of_words = count_vectorizer.fit_transform(sd['cleaned_lyrics'])
````

````python
#print(bag_of_words)
````

````python
simplified_moods = []
for mood in sd['moods']:
    if 'happy' in mood:
        simplified_moods.append('happy')
    elif 'sad' in mood:
        simplified_moods.append('sad')
    else: 
        simplified_moods.append(mood)
len(simplified_moods)
````

````python
se2 = pd.Series(simplified_moods)
sd['simplified_moods'] = se2.values
````

### Picking all the 'happy' & 'sad' songs and Balancing the Data

````python
sd[(sd['simplified_moods'] == 'happy') | (sd['simplified_moods'] == 'sad')].shape
````

````python
sd[(sd['simplified_moods'] == 'happy')].shape
````

````python
#sampling 1249 'happy':
sd_happy = sd[(sd['simplified_moods'] == 'happy')].sample(n=1249, replace=True)
````

````python
sd_happy.shape
````

````python
sd[(sd['simplified_moods'] == 'sad')].shape
````

***Concatenating the songs with lyrics in either 'happy' or 'sad' mood***

````python
happy_or_sad_DF = pd.concat((sd_happy, sd[(sd['simplified_moods'] == 'sad')]), axis=0)
````

````python
happy_or_sad_DF.shape
````

# Bag of Words (BOW)

````python
bag_of_words_happy_sad = count_vectorizer.fit_transform(happy_or_sad_DF['cleaned_lyrics'])
````

````python
bag_of_words_happy_sad.shape
````

## BOW - Logistic Regression

````python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score 

logreg = LogisticRegression()

X = bag_of_words_happy_sad
y = happy_or_sad_DF['simplified_moods']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

logreg.fit(X_train, y_train)
logreg_bow_predictions = logreg.predict(X_test)

print(accuracy_score(y_test, logreg_bow_predictions))
````

***BOW: LogReg ANALYSIS***
Bag of Words technique logistic regression shows 79% accuracy of whether it predicts a song is "happy" or "sad" based on how BOW evaluated the lyrics.


**BOW => LogReg GridSearchCV in search of best params**

````python
from sklearn.model_selection import GridSearchCV
LR_param_grid = {'solver': ['newton-cg', 'sag', 'saga', 'lbfgs'], 'multi_class': ['ovr', 'multinomial']}
````

Code Run: 
```

```
Best parameters returned: {'multi_class': 'ovr', 'solver': 'newton-cg'}

````python
#logreg_grid.best_params_{'multi_class': 'ovr', 'solver': 'newton-cg'}
````

### Optimizing with best params and best features using SelectFromModel

````python
logreg_bow_best = LogisticRegression(multi_class='ovr', solver='newton-cg')
````

````python
logreg_bow_selector = SelectFromModel(logreg_bow_best)
X_new_logreg = logreg_bow_selector.fit_transform(X, y)
````

````python
X_new_logreg.shape
````

````python
from sklearn.metrics import confusion_matrix, classification_report


X = X_new_logreg
y = happy_or_sad_DF['simplified_moods']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

logreg_bow_best.fit(X_train, y_train)
logreg_bow_best_predictions = logreg_bow_best.predict(X_test)

print('accuracy score: ', accuracy_score(y_test, logreg_bow_best_predictions))
print(confusion_matrix(y_test, logreg_bow_best_predictions))
print(classification_report(list(y_test), list(logreg_bow_best_predictions)))
````

### Analysis:
After applying GridSearchCV and SelectFromModel to Bag of Words (SFM chooses 149 words), the accuracy in predicting whether a song is 'happy' or 'sad' increases from 77% to 83%.

```python
import pandas as pd
import numpy as np

songs_with_lyrics_df = pd.read_json('songs_have_lyrics.json')
```

```python
songs_with_lyrics_df.reset_index()
```

```python
moods_list = []
unique_moods = []
for song_mood_list in songs_with_lyrics_df['moods']:
    if len(song_mood_list) > 0:
        moods_list.append(song_mood_list)
        for mood in song_mood_list:
            if mood not in unique_moods:
                unique_moods.append(mood)


print('number of unique moods: ', len(unique_moods))
print(unique_moods)

unique_mood_count = []
for mood in unique_moods:
    count = 0
    for song_mood_list in songs_with_lyrics_df['moods']:
        if mood in song_mood_list:
            count = count + 1
    unique_mood_count.append(count)
unique_mood_count

sum = 0
for mood in unique_mood_count:
    sum = sum + mood
print(sum)

unique_mood_count_dict = {}
for i in range(len(unique_moods)):
    unique_mood_count_dict[unique_moods[i]] = unique_mood_count[i]
print(unique_mood_count_dict)
mood_count = pd.DataFrame.from_dict(data=unique_mood_count_dict, orient='index')
mood_count
```

# Contexts

```python
contexts_list = []
unique_contexts = []
for song_context_list in songs_with_lyrics_df['context']:
    if len(song_context_list) > 0:
        contexts_list.append(song_context_list)
        for context in song_context_list:
            if context not in unique_contexts:
                unique_contexts.append(context)

print(len(contexts_list))
print(len(unique_contexts))
print(unique_contexts)
```

```python
unique_context_count = []
for context in unique_contexts:
    count = 0
    for song_context_list in songs_with_lyrics_df['context']:
        if context in song_context_list:
            count = count + 1
    unique_context_count.append(count)
unique_context_count


unique_contexts_count_dict = {}
for i in range(len(unique_contexts)):
    unique_contexts_count_dict[unique_contexts[i]] = unique_context_count[i]
print(unique_contexts_count_dict)
values_list = unique_contexts_count_dict.values()

context_count = pd.DataFrame.from_dict(data=unique_contexts_count_dict, orient='index')
context_count
```

# Subcontexts

```python
for row in songs_with_lyrics_df.loc[songs_with_lyrics_df.sub_context.isnull(), 'sub_context'].index:
    songs_with_lyrics_df.at[row, 'sub_context'] = []

songs_by_subcontext_series = songs_with_lyrics_df['sub_context'].tolist()

unique_subcontexts = []
for song_subcontext_list in songs_with_lyrics_df['sub_context']:
    for sub_context in song_subcontext_list:
        if sub_context not in unique_subcontexts:
            unique_subcontexts.append(sub_context)

#print(len(songs_by_subcontext_series)) 20931 == # of songs w/ lyrics
#print(len(unique_subcontexts)) #62
print(unique_subcontexts)
```

```python
unique_subcontext_count = []
for subcontext in unique_subcontexts:
    count = 0
    for song_subcontext_list in songs_with_lyrics_df['sub_context']:
        if subcontext in song_subcontext_list:
            count = count + 1
    unique_subcontext_count.append(count)
unique_subcontext_count


unique_subcontexts_count_dict = {}
for i in range(len(unique_subcontexts)):
    unique_subcontexts_count_dict[unique_subcontexts[i]] = unique_subcontext_count[i]
print(unique_subcontexts_count_dict)
values_list = unique_subcontexts_count_dict.values()

subcontext_count = pd.DataFrame.from_dict(data=unique_subcontexts_count_dict, orient='index')
subcontext_count
```

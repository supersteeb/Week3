```python
import pandas as pd
import numpy as np
```

```python
songs_df = pd.read_json('MasterSongList.json')
```

```python
songs_df.head()
```

```python
songs_df.loc[:, 'genres'] = songs_df['genres'].apply(''.join)

def consolidateGenre(genre):
    if (len(genre) > 0):
        return genre.split(':')[0]
    else:
        return genre
    
songs_df.loc[:, 'genres'] = songs_df['genres'].apply(consolidateGenre)
```

```python
audio_feature_list = []
for song_audio_feature in songs_df['audio_features']:
    audio_feature_list.append(song_audio_feature)
```

```python
audio_feature_headers = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']
```

```python
df_features = pd.DataFrame(audio_feature_list, columns=audio_feature_headers)
```

```python
df_features['genres'] = songs_df['genres']
```

```python
df_features['genres'].unique()
```

```python
df_features['genres'].replace('',np.NaN, inplace=True)
```

```python
df_features.head()
```

```python
df_features.dropna(inplace=True)
```

```python
df_features.shape
```

```python
df_features_rap = df_features[df_features['genres'] == 'rap']
```

```python
df_features_rap.shape
```

```python
df_features_rap['genres_simplified'] = 'rap'
```

# make sure sample is same amount of the genre amount (balancing data) 
('classical' => 2452)

```python
df_features_not_rap = df_features[df_features['genres'] != 'rap'].sample(n=2452)
```

```python
df_features_not_rap['genres_simplified'] = 'not rap'
```

```python
df_features_RapNotRap = pd.concat([df_features_rap, df_features_not_rap], axis=0)
```

```python
df_features_RapNotRap.reset_index(inplace=True)
```

```python
df_features_RapNotRap
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df_features_RapNotRap.drop(['genres', 'genres_simplified'], axis=1)
y = df_features_RapNotRap['genres_simplified']
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

rfc = RandomForestClassifier()
selector = SelectKBest()
knn = KNeighborsClassifier()
svc = SVC(probability=True)
logReg = LogisticRegression()
```

# importing genreRapNotRap_votingClassifier

```python
genreRapNotRap_votingClassifier = VotingClassifier([('<knn>', knn), ('<svc>', svc), ('<logReg>', logReg), ('<rfc>', rfc)], voting='soft')
```

```python
af_RapNotRap = genreRapNotRap_votingClassifier.fit(X_train, y_train)
```

```python
y_predictions_RapNotRap = genreRapNotRap_votingClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_predictions_RapNotRap)) #84.7
```

```python
import pandas as pd
import numpy as np
```

```python
songs_df = pd.read_json('MasterSongList.json')
```

```python
songs_df.head()
```

```python
songs_df.loc[:, 'genres'] = songs_df['genres'].apply(''.join)

def consolidateGenre(genre):
    if (len(genre) > 0):
        return genre.split(':')[0]
    else:
        return genre
    
songs_df.loc[:, 'genres'] = songs_df['genres'].apply(consolidateGenre)
```

```python
audio_feature_list = []
for song_audio_feature in songs_df['audio_features']:
    audio_feature_list.append(song_audio_feature)
```

```python
audio_feature_headers = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']
```

```python
df_features = pd.DataFrame(audio_feature_list, columns=audio_feature_headers)
```

```python
df_features['genres'] = songs_df['genres']
```

```python
df_features['genres'].unique()
```

```python
df_features['genres'].replace('',np.NaN, inplace=True)
```

```python
df_features.head()
```

```python
df_features.dropna(inplace=True)
```

```python
df_features.shape
```

```python
df_features_rnb = df_features[df_features['genres'] == 'r&b']
```

```python
df_features_rnb.shape
```

```python
df_features_rnb['genres_simplified'] = 'r&b'
```

# make sure sample is same amount of the genre amount (balancing data)

```python
df_features_not_rnb = df_features[df_features['genres'] != 'r&b'].sample(n=2344)
```

```python
df_features_not_rnb['genres_simplified'] = 'not r&b'
```

```python
df_features_RnbNotRnb = pd.concat([df_features_rnb, df_features_not_rnb], axis=0)
```

```python
df_features_RnbNotRnb.reset_index(inplace=True)
```

```python
df_features_RnbNotRnb
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df_features_RnbNotRnb.drop(['genres', 'genres_simplified'], axis=1)
y = df_features_RnbNotRnb['genres_simplified']
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

rfc = RandomForestClassifier()
selector = SelectKBest()
knn = KNeighborsClassifier()
svc = SVC()
logReg = LogisticRegression()
```

# importing genreRnbNotRnb_votingClassifier

```python
genreRnbNotRnb_votingClassifier = VotingClassifier([('<knn>', knn), ('<svc>', svc), ('<logReg>', logReg), ('<rfc>', rfc)])
```

```python
af_RnbNotRnb = genreRnbNotRnb_votingClassifier.fit(X_train, y_train)
```

```python
y_predictions_RnbNotRnb = genreRnbNotRnb_votingClassifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test, y_predictions_RnbNotRnb)) #78.5%
```
